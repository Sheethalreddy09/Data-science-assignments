```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('sales_data_with_discounts.csv')
df
```

```python
for column in df.select_dtypes(include='number').columns:
  print(f"{column}")
  print(f"Mean : {df[column].mean()}")
  print(f"Median : {df[column].median()}")
  print(f"Mode : {df[column].mode()}")
  print(f"Standard Deviation : {df[column].std()}")
  print("-" * 40)
```

```python
from typing import OrderedDict
from typing_extensions import dataclass_transform
from collections.abc import ValuesView
#Brief Interpretation of the Above statistics
Mean : Mean is known as average value and it is calculated as
               Mean = Sum of values / Total no of Values
Median : Median is defined as middle most value of the data. Median is calculated by sorting the data in ascending or descending.

Mode : Mode is most frequentlu occured value is called mode of the data.

Standard Deviation : Standard deviation is calculated by Squares of the differnce between values and the mean by total no of values.


```

```python
Data Visualization
●	Objective: To visualize the distribution and relationship of numerical and categorical variables in the dataset.
●	Histograms:
	Plot histograms for each numerical column.
	Analyze the distribution (e.g., skewness, presence of outliers) and provide inferences.
●	Box Plots:
	Create boxplots for numerical variables to identify outliers and the interquartile range.
	Discuss any findings, such as extreme values or unusual distributions.
●	Bar Chart Analysis for Categorical Column:
	Identify categorical columns in the dataset.
	Create bar charts to visualize the frequency or count of each category.
	Analyze the distribution of categories and provide insights.
```

```python
#Histograms PLots
for column in df.select_dtypes(include='number').columns:
  plt.figure(figsize=(10,5))
  plt.hist(df[column],bins=20,alpha=0.9)
  plt.title(f" Histogram of {column}")
  plt.xlabel(column)
  plt.ylabel("Frequency")
  plt.show()
```

```python
#Brief Interpretation about Histogram by each colum
1. The Histogram of Volume shows that most values lies between 0 to 5. and The distribution is positive skewed because the mean > median > mode. the spread indicates high variability.

2.The histogram of AvgPrice colums show that the most values lies in the range 0 to 1000.it has descrete data.

3. The histogram of Total Sale Value shows that most of the sales range is between 0 to 10000. It is positive skewed

4. Histogram plot of Discount rate it has a negative skewed. most values are in range from 15 to 20

5. The Discount amount has a positive skewness. And it has continous data.

6. The Net Sales Value is also a positive skewed. and the spread indicates the high variability.
```

```python
#box plot for numerical columns
for column in df.select_dtypes(include='number').columns:
  plt.figure(figsize=(10,5))
  plt.boxplot(df[column])
  plt.title(f"Box plot of {column}")
  plt.show()
```

```python
#interepretation of box plot
1. the box plot of volume has outliers from 25 to 30

2. the box plot of Avg price has outliers from 50000 to 60000

3. the box plot of Total Sales Value has outliers from 190000 to 200000

4. the box plot of Discount amount  has outliers from 20000 to 25000

5. the box plot of Net sales Valuehas outliers from 160000 to 175000

6. The box plot of Discount rate has outliers from 6 to 8 and it is lower extreme box plot

```
