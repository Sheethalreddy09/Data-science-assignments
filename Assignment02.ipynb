```python
6.	Given a 'marks' variable, print the grade:
- 90 and above: 'A'
- 75–89: 'B'
- 50–74: 'C'
- Below 50: 'Fail'
```

```python
#code for grade
marks=int(input("enter your marks :"))
if(marks>=90):
  print("Grade A")
elif(marks>=75 and marks<=89):
  print("Grade B")
elif(marks>=50 and marks<=74):
  print("Grade C")
else:
  print("Fail")
```

```python
one_dime=np.array([1,2,3,4,5])
print("One Dimension array")
print("Shape :",one_dime.shape)
print("Dimensions :",one_dime.ndim)
print(one_dime)
```

```python
two_dim=np.arange(10,70,10).reshape(2,3)
print("Two Dimensional Array")
print("Shape :",two_dim.shape)
print("Dimension :",two_dim.ndim)
print(two_dim)
```

```python
11.	Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:
- Use .info(), .describe()
- Select 'Name' and 'City' columns
- Drop 'City' column
- Fill any missing values in 'Salary' column with the mean
- Remove any duplicate rows
```

```python
import pandas as pd
import numpy as np
data = {
    "Name": ["Ravi", "Anita", "Kiran", "Ravi", "Vijay",
             "Meena", "Arjun", "Priya", "Rahul", "Neha"],
    "Age": [25, 28, 22, 30, 35,
            27, 24, 29, 26, 23],
    "City": ["Delhi", "Mumbai", "Bangalore", "Hyderabad", "Chennai",
             "Pune", "Kolkata", "Jaipur", "Ahmedabad", "Lucknow"],
    "Salary": [50000, 60000, np.nan, 70000, 80000,
               55000, 48000, np.nan, 52000, 47000]
}
df = pd.DataFrame(data)
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df[['Name','City']]
```

```python
df.drop('City',axis=1)
```

```python
mean_salary=df['Salary'].mean()
df['Salary'].fillna(mean_salary,inplace=True)
df
```

```python
dup=df.drop_duplicates('Name')
dup
```
