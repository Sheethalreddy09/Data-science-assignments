```python
#by using different thresholds
def recommend_by_threshold(title, threshold=0.3):
    idx = new_data[new_data['name'] == title].index[0]

    # get similarity scores
    sim_scores = list(enumerate(cosine_sim[idx]))

    # filter scores based on threshold
    filtered = [i for i in sim_scores if i[1] >= threshold and i[0] != idx]

    # sort from highest to lowest
    filtered = sorted(filtered, key=lambda x: x[1], reverse=True)

    # get anime indices
    anime_indices = [i[0] for i in filtered]

    return data[['name', 'genre', 'rating']].iloc[anime_indices]
```
