```python
from sklearn.preprocessing import MinMaxScaler
X=data.drop('Y',axis=1)
y=data['Y']
scaler=MinMaxScaler()
X_scaled=scaler.fit_transform(X)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y_encoded= encoder.fit_transform(y)
```

2. Model Implementation
●	Construct a basic ANN model using your chosen high-level neural network library. Ensure your model includes at least one hidden layer.
●	Divide the dataset into training and test sets.
●	Train your model on the training set and then use it to make predictions on the test set.


```python
#train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y_encoded,test_size=0.2,random_state=42,stratify=y_encoded)
```

```python
print(X_train.shape,X_test.shape)
```

```python
#model building
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
num_classes=len(encoder.classes_)
```

```python
num_classes
```

```python
model = Sequential([
    Dense(128, activation='relu', input_shape=(60,)),   # 1 hidden layer
    Dense(64, activation='relu'),                       # Optional 2nd hidden layer
    Dense(num_classes, activation='softmax')            # Output layer
])
```

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model.summary()
```

```python
#fitting the model
model.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=20,
    batch_size=32,
    verbose=1
)
```

```python
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print("Test Accuracy:", test_accuracy)
```

```python
y_pred = model.predict(X_test)
y_pred_classes = y_pred.argmax(axis=1)

print(y_pred_classes[:10])
```

3. Hyperparameter Tuning
●	Modify various hyperparameters, such as the number of hidden layers, neurons per hidden layer, activation functions, and learning rate, to observe their impact on model performance.
●	Adopt a structured approach like grid search or random search for hyperparameter tuning, documenting your methodology thoroughly.


```python
!pip install keras-tuner
```

```python
import keras_tuner as kt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam, SGD

def build_model(hp):

    model = Sequential()

    # Choose number of hidden layers (1 to 3)
    for i in range(hp.Int("num_layers", 1, 3)):
        model.add(Dense(
            units=hp.Choice("units_"+str(i), [32, 64, 128, 256]),
            activation=hp.Choice("activation", ["relu", "tanh"])
        ))

        # optional dropout
        if hp.Boolean("dropout_"+str(i)):
            model.add(Dropout(0.3))

    # Output layer
    model.add(Dense(num_classes, activation="softmax"))

    # Optimizer & learning rate
    lr = hp.Choice("learning_rate", [1e-4, 3e-4, 1e-3, 3e-3])
    opt = Adam(learning_rate=lr)

    model.compile(
        optimizer=opt,
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"]
    )

    return model
```

```python
tuner = kt.RandomSearch(
    build_model,
    objective="val_accuracy",
    max_trials=10,          # number of hyperparameter combinations
    executions_per_trial=1,
    directory="ann_tuning",
    project_name="alphabet_model"
)
```

```python
tuner.search(
    X_train, y_train,
    epochs=20,
    validation_split=0.2,
    verbose=2
)
```

```python
best_model = tuner.get_best_models(num_models=1)[0]
best_model.summary()
```

```python
test_loss, test_acc = best_model.evaluate(X_test, y_test)
print("Test Accuracy:", test_acc)
```

4. Evaluation
●	Employ suitable metrics such as accuracy, precision, recall, and F1-score to evaluate your model's performance.
●	Discuss the performance differences between the model with default hyperparameters and the tuned model, emphasizing the effects of hyperparameter tuning


```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix

# Predictions
y_pred_base = model.predict(X_test)
y_pred_base = y_pred_base.argmax(axis=1)

print("Baseline Model Classification Report:")
print(classification_report(y_test, y_pred_base, target_names=encoder.classes_))

print("Baseline Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_base))
```

```python
# Predictions from tuned model
y_pred_tuned = best_model.predict(X_test)
y_pred_tuned = y_pred_tuned.argmax(axis=1)

print("Tuned Model Classification Report:")
print(classification_report(y_test, y_pred_tuned, target_names=encoder.classes_))

print("Tuned Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_tuned))
```
