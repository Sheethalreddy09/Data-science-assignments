```python
#1st question
n = int(input())
if n<=1:
    print("not a prime number")
else:
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            print('not a prime number')
            break
    else:
            print('prime number')
```

```python
#4th question
days = 4
tests = 3
medicine_charges = 2400
room_charges = 2000*days
consultation_fees = 1500
test_charges = 300*tests
total_bill = room_charges+consultation_fees+test_charges+medicine_charges
if total_bill>10000:
    discount = total_bill * 0.10
    final_bill = total_bill - discount
else:
    final_bill = total_bill

print("Final Bill Amount =", final_bill)
```
